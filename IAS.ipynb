```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.callbacks import EarlyStopping

x_train_st, x_test_st, y_train_st, y_test_st = train_test_split(
    train_data, label_data, test_size=0.25, random_state=42)
model = Sequential()
model.add(Dense(62, input_dim=train_data.shape[1], activation='relu'))
model.add(Dense(40, input_dim=train_data.shape[1], activation='relu'))
model.add(Dense(10, input_dim=train_data.shape[1], activation='relu'))
model.add(Dense(1, kernel_initializer='normal'))
model.add(Dense(label_data.shape[1],activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam')
monitor = EarlyStopping(monitor='val_loss', min_delta=1e-3, 
                        patience=5, verbose=1, mode='auto')
model.fit(x_train_st,y_train_st,validation_data=(x_test_st,y_test_st),
          callbacks=[monitor],verbose=2,epochs=500)
```
